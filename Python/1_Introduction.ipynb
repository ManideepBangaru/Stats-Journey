$\sf \Large \textbf{1. Introduction to Statistics}$

* Statistics is a way of thinking
* To minimize the errors we use a framework which allows us to provide solutions in a structured way

$\sf \normalsize \textbf{The DCOVA Framework : }$
* **Define :** Define the data you want to study to solve a problem
* **Collect :** Collect the data using appropriate sources
* **Organize :** Organize the data collected into tables
* **Visualize :** Visulaize the data collected by developing plots
* **Analyze :** Analyze the data to reach conclusions

$\sf \normalsize \textbf{Nature of Statistical law : }$\
\
**Law of Statistical regularity** states that from a very large population, if you select a moderate large sample, then the sample is highly likely to have the characteristics of the population.

**Law of inertial of large numbers** is by principles of law of statistical regularity states that if the sample is bigger in size, then the study yields the results that of population

$\sf \large \textbf{Defining data}$

Data can be divided into two types :
* **Categorical**  $\rightarrow$ Represented in categories
* **Numerical**    $\rightarrow$ Counted or Measured quantities
    * **Discrete** : Which arise from a process of counting, ex:- Number of Tvs sold in a store
    * **Continuous** : Which arise from a process of measuring, ex:- Body temperature of a person

$\sf \normalsize \textbf{Measurement scales : }$\
Above mentioned types of data can be fallen into any of the below mentioned scales.
* **Numeric - Interval scale :** Which expresses a difference between measurements that do not include a true zero point
* **Numeric - Ratio scale :** Which has a true zero point, ex:- one item is twice as expensive as other
* **Categorical - Nominal :** Which doesn't have any order or ranking
* **Categorical - Ordinal :** Which has an order or ranking

$\sf \large \textbf{Collecting data}$

### Population and Samples
* **Population :** A population contains all the items or individuals of interest that you seek to study
* **Sample :** A sample contains only a proportion of a population of interest

### Data Sources
* Capturing data generated by ongoing business activities
* Distributing data compiled by an organization or individual
* Compiling the responses from a survey
* Conducting a designed experiment and recording the outcomes of the experiment
* Conducting an observational study and recording the results of the study

#### Primary data sources
When you perform the activity that collects the data, you are using a primary data source

#### Secondary data sources
When the data collection part is done by someone else, you are using a secondary data source

#### Internal Source
Set of internal data if it is obtainable from well within an organization and relates to organization’s operations
Ex:- Computerized files containing sales figures, financial data,operating information and so on

#### External Source
Set of data that is collected outside the organization. This type of data may be available in the published financial periodicals or it may be stored in an internal computer data bank accessible  by an on-line computer terminal


### Data Structures

* **Univariate Data**\
Univariate data sets have a single piece of information recorded for each item.\
**Ex:-** Statistical analysis of data collected regarding the income level through a marketing survey would reveal the distribution of incomes, specific income level, variation in the income level and the number of people within any given range of income

* **Bivariate Data**\
Bivariate (two variables) data sets have exactly two pieces of information recorded for every item. Application of statistical analysis would reveal the relationship between the two variables under study.

* **Multivariate Data**\
Multivariate data (Minimum of three variables) sets have a minimum of three pieces of information recorded for every item. Statistical analysis can be applied to study the interrelationship between all the variables.


### Data Collection methods

* **Observation method** (difficult to produce accurate data)\
The person who collects the data asks no questions, but he observes carefully the phenomenon under consideration, and records the essential data.

* **Questionnaire method** (more accurate and reliable)\
The researcher designs a questionnaire which contains all the relevant questions needed for the study.

* **Mail questionnaire method** (costly one)\
The questionnaire is mailed to the respondent’s residential or official address.

* **Telephonic interview method** (less expensive, somewhat accurate)\
The researcher asks the relevant questions to the respondents over the telephone.

* **Census method**\
When the researcher collects the data from each and every individual of the population
 * **Adv -** complete information, accurate
 * **Disadv -** lot of time and huge money

* **Sampling methods**\
When you collect data by selecting a sample, you begin by defining the frame. The frame is a complete or partial listing of the items that make up the population from which the sample will be selected.
